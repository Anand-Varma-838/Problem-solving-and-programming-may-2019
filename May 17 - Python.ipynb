```python
def addRowCSV(filename,row):
    df = readCSV(filename)
    df.loc[len(df)] = row
    df.to_csv(filename,index=False)
    return df

r=['04000US08','Las Vegas',54637,53456,54967,52234,54342,52455,56234,58235,56546]
addRowCSV('DataFiles/Income.csv',r)
```

### Constraints 

### Test Cases

```python
def updateRowCSV(filename,rowindex,rowdata):
    df = readCSV(filename)
    df.loc[rowindex] = rowdata
    df.to_csv(filename,index=False)
    return df

rowdata = ['04000US10','Manhattan',54635,52456,54867,51234,54352,56455,56274,52235,56046]
updateRowCSV('DataFiles/Income.csv',6,rowdata)
```

### Constraints 

### Test Cases

```python
def deleteRowCSV(filename,rowindex):
    df = readCSV(filename)
    df = df.drop(rowindex)
    df.to_csv(filename,index=False)
    return df
```

```python
filename = 'DataFiles/Income.csv'
rowindex=7
deleteRowCSV(filename,rowindex)
```
